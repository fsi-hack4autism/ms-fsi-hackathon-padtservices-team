```
def parse_time(s):
    if not isinstance(s, str): 
      return 0
    week, day, hour, minu, sec = s.split(':')
    try:
        week=int(week)
        day=int(day)
        hour = int(hour)
        minu = int(minu)
        sec = int(sec)
    except ValueError:
        # handle errors here, but this isn't a bad default to ignore errors
        return 0
    return  week*7*24*60*60 + day*24*60*60 + hour * 60 * 60 + min * 60 + sec

```

```
df_initial_analysis['sessionPeriod'] = df_initial_analysis['sessionPeriod'].fillna(0, inplace=True)	
df_initial_analysis[ 'sessionPeriod' ] = df_initial_analysis[ 'sessionPeriod' ].apply(parse_time)
```

```
def label_goal_domain(s):
  if s == "Communication": 
    return 0
  elif  s == "Language":
    return 1
  elif  s == "Adaptive":
    return 2
  else: 
    return 3
```

```
df_initial_analysis['GoalDomain'] = df_initial_analysis['GoalDomain'].apply(label_goal_domain)
```
